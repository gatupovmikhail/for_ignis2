```python
def reformat(st):
    fl = 0
    if '(' in st:
        st = st.split('(')[0]
        fl = 1
    sttir = st.split('-')
    if not('Finish' in st):
        namem = sttir[0].strip().split('_')[1:]
        name = '_'.join(namem)
    else:
        name = 'Finish'
    if fl == 1:
        name+='*'
    t = int(sttir[1].split(':')[0])*6000 + int(sttir[1].split(':')[1])*100 + int(sttir[1].split(':')[2])
    return(name,t)
reformat('14_white - 01:02:03')
```
